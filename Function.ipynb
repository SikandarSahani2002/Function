#Theory Questions:

 1. What is the difference between a function and a method in Python?
- A function is a block of code that works independently and can be called directly.<br>
A method is similar to a function but is always associated with an object, meaning it works on data stored inside a class.<br>
Example<br>
Function > independent tool<br>
Method > tool inside an object

 2. Explain the concept of function arguments and parameters in Python.
- Parameters are the names written in the function definition.<br>
Arguments are the actual values you pass when calling the function.<br>
- Example<br>
- Parameters = empty placeholders inside the function
- Arguments = actual data you send to the function

 3. What are the different ways to define and call a function in Python?
- We can define a function normally using the def keyword and call it by its name.
- We can also create functions with arguments, default arguments, or use a short unnamed function called a lambda.
- In all cases, you call the function by writing its name followed by brackets.<br>
**Example**<br>
Like calling a friend by name — sometimes you give details (arguments), sometimes you don’t.


 4. What is the purpose of the 'return' statement in a Python function?
 - The return statement sends a value back from a function to the place where it was called.<br>
It helps the function give an output instead of just printing something.<br>
**Example:**<br>
Like asking someone for the price of a product — they return the final amount to you.


 5. What are iterators in Python and how do they differ from iterables?
 - An iterable is any object we can loop through (like a list or string).<br>
An iterator is the object that actually gives one value at a time when looping.<br>
** Example:**<br>
An iterable is like a book, and an iterator is like reading the book page by page.


 6. Explain the concept of generators in Python and how they are defined.
- Generators are special functions that produce values one at a time instead of creating the whole list in memory.<br>
- They are defined using the 'yield' keyword instead of return.<br>
**Example:**<br>
A generator works like a water tap — it gives water flow by flow, not all at once.

 7. What are the advantages of using generators over regular functions?
- Generators use less memory because they produce values one by one.
They are faster for large data and do not store the entire result at once.
They make looping more efficient.<br>
**Example:**<br>
A generator is like getting fruits one by one from a shop instead of carrying a whole heavy box at once.

 8. What is a lambda function in Python and when is it typically used?
- A lambda function is a small anonymous function defined in one line using the lambda keyword.
It is typically used for simple operations where a full function is not needed.<br>
**Example:**<br>
Like quickly writing a shortcut note instead of a full letter — useful for small tasks.

 9. Explain the purpose and usage of the `map()` function in Python.
- The map() function applies a given function to all items of an iterable and returns a new iterable with the results.
It is used to perform the same operation on multiple items quickly.<br>
**Example:**<br>
Like coating all cookies with chocolate at once instead of one by one manually.

 10. What is the difference between 'map()', 'reduce()', and filter()` functions in Python?
- map() applies a function to all items of an iterable and returns a new iterable.
- filter() selects items from an iterable that meet a condition.
- reduce() (from functools) combines all items into a single value using a function.<br>
***Example:***<br>
- map() → glaze all cookies
- filter() → pick only chocolate cookies
- reduce() → stack all cookies into one tower

```python
''' 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list'''

#function
def sum_even(num):
    total = 0
    for n in num:
        if n % 2 == 0:
            total += n
    print("The sum of totaL Even number is",total)

sum_even([1,2,3,4,5,6,7,8,9])


```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
#each number

def sqr_lst(num):
    result = []
    for n in num:
        result.append(n*n)
    print("The new list is ",result)

sqr_lst([1,2,3,4,5])
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def pwr_of_2(n):
    for i in range(n + 1):
        yield 2 ** i

# Example usage:
for value in pwr_of_2(12):
    print(value)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file(filename):
    try:
        with open(filename, "r") as file:
            for line in file:
                yield line.strip()#strip() for removing the extra spaces
    except FileNotFoundError:
        yield "File not found!"
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# Sample list of tuples
data = [(1, 9), (7, 2), (3, 11), (6, 1)]

# Sort using lambda
sorted_data = sorted(data, key=lambda x:x[1])

print(sorted_data)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# list of temperatures in Celsius
celsius = [0, 20, 37, 100]

# Function to convert Celsius to Fahrenheit
def celsius_to_fehrenheit(c):
    return (c * 9/5) + 32

# Using map to convert all values
fahrenheit = list(map(celsius_to_fehrenheit, celsius))

print(fahrenheit)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

# Sample string
text =input("Enter a string :").lower()

# Function to check if a character is not a vowel
def not_vowel(char):
    return char.lower() not in "aeiou"

# Using filter to remove vowels
result = " ".join(filter(not_vowel, text))

print(result)
```
